```python
# EDA uing Bivariate and Multivariate
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
titanic = pd.read_csv(r'Titanic-Dataset.csv')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
# print(tips.head())
# print(iris.head())

# Scattrt ploy
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'], style=tips['smoker'], size=tips['size'])
plt.show()
```

```python
print(titanic.head())
sns.barplot (x='Pclass', y='Age', data=titanic)
plt.show()
```

```python
sns.barplot (x='Pclass', y='Fare', data=titanic, hue='Sex')
plt.show()
```

```python
sns.boxplot(data=titanic, x='Sex', y='Age', hue='Survived')
plt.show() 
```

```python
sns.distplot(titanic['Age'])
plt.show()
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
plt.show()
```

```python

# Heaetmap
print(pd.crosstab(titanic['Pclass'], titanic['Survived']))
Vitualization by heatmap
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
plt.show()
```

```python
print(titanic.groupby('Pclass')['Survived'].mean()*100)
print(titanic.groupby('Sex')['Survived'].mean()*100)

sns.heatmap(pd.crosstab(titanic['Sex'],titanic['Survived']),annot=True)
plt.show()
```

```python
Cluster Mapping
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))      # which values behavior is closer
plt.show()                  
```

```python
pairplot: collection of scatterplot
print(iris.head())
sns.pairplot(iris, hue='species')               # important when there are moore numerical values
plt.show()
```

```python
lineplot 
print(flights.head())
new = flights.groupby('year')['passengers'].sum()
print("New****************************",new)
sns.lineplot(new)
plt.show()
```

```python
new2= flights.pivot_table(values='passengers', index='month', columns='year')
print(new2)
sns.heatmap(new2)
plt.show()
sns.clustermap(new2)
plt.show()
```
