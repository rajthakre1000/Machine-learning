## 1. Ordinal encoding
###  data(x) contain ordional data: apply ordinal encoding,
### target (y) contain ordinal : apply Lable encoding


```python
import pandas as pd
data = pd.read_csv(r'customer.csv')
# print(data.head())
```

```python

data = data.iloc[:,2:]
print(data.sample(5))
from sklearn.model_selection import train_test_split

```

```python
x_train, x_test, y_train, y_test = train_test_split(data.iloc[:,0:2], data.iloc[:,-1], test_size=.2)
# print(x_train.shape, x_test.shape)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
oe.fit(x_train)
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
print(x_train)

print(oe.categories_)

# Lable encoding
# print(y_train)
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
lb.fit(y_train)
y_train = lb.transform(y_train)
y_test = lb.transform(y_test)
print(y_train)
print()
```
