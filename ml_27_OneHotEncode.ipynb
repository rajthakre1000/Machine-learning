```python
#* Dummy Variable trap
# we can remove any one colum from n columns (Multicollinearity)
import pandas as pd
import numpy as np
```

```python

# print(data1, data1.shape)   #(8128,12)
#  Dummy Variable trap
data2 = pd.get_dummies(data=data, columns= ['owner', 'fuel'], drop_first=True)
# print(data2.shape)      # from both columns one one columns reduced (8128,10)
print(data2.head(10))
```

```python
# OneHotEncoding using Sklean 
from sklearn.model_selection import train_test_split
x_train, x_test, Y_train, y_text = train_test_split(data.iloc[:,0:4], data.iloc[:,-1] ,test_size=.2)
print(x_train.head())
```

```python

from sklearn.preprocessing import OneHotEncoder
obj = OneHotEncoder(drop='first',sparse_output=False)           # by default produces :Sparse Matrics, use sparse_output= perameter
#print(x_train_new.toarray())            # Sparse matrics produced by OneHotEncoder
x_train_new = obj.fit_transform(x_train[['fuel', 'owner']])
x_test_new = obj.transform(x_test[['fuel', 'owner']])
print(x_train_new)
```

```python
final_array = np.hstack((x_train[['brand','km_driven']].values, x_train_new))               # work done
print(final_array)
```
