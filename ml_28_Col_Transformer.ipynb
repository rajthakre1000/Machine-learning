```python
import pandas as pd
data = pd.read_csv(r'covid_toy.csv')
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python

data = pd.read_csv(r'covid_toy.csv')

print(data.head())
print(data['cough'].value_counts())
print(data['city'].value_counts())
print(data.isnull().sum())
```

```python

x_train, x_test, y_train, y_test = train_test_split(data.drop(columns='has_covid'), data['has_covid'], test_size=.2)
# print(x_train)
# print(y_test)
```

```python

from sklearn.compose import ColumnTransformer
transform = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(), ['fever']),
    ('tnf2', OneHotEncoder(sparse_output=False, drop='first'), ['city', 'gender']),
    ('tnf3', OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough'])
], remainder='passthrough')
```

```python
print("**************************************************************************")
transform.fit_transform(x_train)
transform.fit_transform(x_test)
```
