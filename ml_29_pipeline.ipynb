```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder 
from sklearn.preprocessing import minmax_scale
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r'Titanic-Dataset.csv')
print(df.head())
df = df.drop(columns=['PassengerId', 'Name','Ticket','Cabin'])
print(df.head())
```

```python

x_train, x_test, y_train, y_test = train_test_split(df.drop(columns='Survived'), df['Survived'], test_size= .2)
print(x_train.shape)

# print(df.isnull().sum())            #age = 177 numm value
```

```python

# SimpuleImputer:  Embarked, Age
si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')
```

```python

x_train_age = si_age.fit_transform(x_train[['Age']])
x_train_embarked = si_embarked.fit_transform(x_train[['Embarked']])
 
x_test_age = si_age.transform(x_test[['Age']])
x_test_embarked = si_embarked.transform(x_test[['Embarked']])

# print(x_train_age)
```

```python

# OneHotEncoder Age and Embarked
hot_sex = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
hot_embarked = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')
```

```python

x_train_sex= hot_sex.fit_transform(x_train[['Sex']])
x_test_sex= hot_sex.transform(x_test[['Sex']])
```

```python

x_train_embarked= hot_embarked.fit_transform(x_train_embarked)
x_test_embarked= hot_embarked.transform(x_test_embarked)
# print(x_test_embarked)
# print(x_test_sex)
```

```python

x_train_rem = x_train.drop(columns=['Sex','Age','Embarked'])
x_test_rem = x_test.drop(columns=['Sex','Age','Embarked'])
# print(x_train_rem)
x_train_trantransform= np.concatenate((x_train_rem, x_train_age, x_train_sex, x_train_embarked), axis=1)
x_test_trantransform= np.concatenate((x_test_rem, x_test_age, x_test_sex, x_test_embarked), axis=1)
print(x_train_trantransform)
```

```python

# Training model
clf = DecisionTreeClassifier()
clf.fit(x_train_trantransform, y_train)
y_pred = clf.predict(x_test_trantransform)
print(y_pred)
```

```python

print(accuracy_score(y_test, y_pred)*100)
```
