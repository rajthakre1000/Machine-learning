# 1. Discretization
1. Discretization is process of transforming continuous variables into discrete variabes by contiguous intervals that span the range of the variable' value. bin is alternative name of interval.
2. To handle Outliers, To improve the value spread
3. (max - min)/bins = number of intervals

Types of Discretization : 
1. Equal Width (max-min)/bins = interval
2. Equal Frequency (euqal frequency of elements in each interval)
3. KMean binning (Clusturing algorithm : N Dimention clusters partitio)


K-means algorithm to a single feature:
1. Specify K: You first decide on the number of bins (k) you want to create.
2. Initialize Centroids: The algorithm randomly selects k initial center points (centroids) from the range of the data.
3. Assign Data Points: Each data point is assigned to the nearest centroid using a distance measure, typically the Euclidean distance. This forms initial clusters.
4. Update Centroids: The mean (centroid) of all data points within each cluster is recalculated. The centroid position shifts to the center of its assigned points.
5. Iterate: Steps 3 and 4 are repeated until the centroids no longer move significantly or the maximum number of iterations is reached (convergence).
6. Define Bin Boundaries: The final positions of the converged centroids (or the natural breaks between clusters) define the optimal bin boundaries.
