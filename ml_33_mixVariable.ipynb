# Working
1. convert data into two diffrent column storing catagorical and numerical data.


```python
import pandas as pd
import numpy as np 

df = pd.read_csv(r'c:\Users\rajku\Downloads\titanic.csv')

print(df.head())
```

```python
print(df['number'].unique())
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passenger travelling with')
```

```python
# creating new column
df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
# coerce - invalid values become NaN (1,2,3,A,6) -> (1,2,3,naN,6).
```

```python
df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan )
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')   #capture numerical ppart
df['cabin_categorical'] = df['Cabin'].str[0]         #capture first letter
```
