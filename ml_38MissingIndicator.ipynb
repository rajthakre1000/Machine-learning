# Missing Indicator 
Making new column for showing missing values, (true/ false)

```python
# sometimes model learns how to diffrentiate
# btw missing and non missing values
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator, SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r'c:\Users\rajku\Downloads\train (1).csv', usecols=['Age', 'Fare', 'Survived'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
x_train.head()
```

```python
s = SimpleImputer()
x_train_imputed = s.fit_transform(x_train)
x_test_imputed = s.transform(x_test)
```

```python
x_train_imputed
```

```python
clf = LogisticRegression()
clf.fit(x_train_imputed, y_train)
```

```python
y_pred = clf.predict(x_test_imputed)
print(accuracy_score(y_test, y_pred))
```

```python
mi = MissingIndicator()
mi.fit(x_train)
x_train_miss = mi.transform(x_train)
x_test_miss = mi.transform(x_test)
```

```python
x_train['Age_NA'] = x_train_miss
x_test['Age_NA'] = x_test_miss
```

```python
clf2 = LogisticRegression()
clf2.fit(x_train_miss, y_train)
# print(x_train)
```

```python
y_pred = clf2.predict(x_test_miss)
print(accuracy_score(y_pred, y_test))
```

```python
si = SimpleImputer(add_indicator=True)          # inbuild perameter for missing value indicator
s = SimpleImputer()
x_train_imputed = si.fit_transform(x_train)
x_test_imputed = si.transform(x_test)

clf = LogisticRegression()
clf.fit(x_train_imputed, y_train)

y_pred = clf.predict(x_test_imputed)
print(accuracy_score(y_test, y_pred))
    
```
