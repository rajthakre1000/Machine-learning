# KNN imputation : multivariate imputation, in order to fill a missing value.

fill the value , with the row which is similar to this one the most, (by euclidean distance)

2 step process.
1. find k nearest nabour.
2. find the value.
nan_euclidean_distance().

Finds euclidean distance of each row, col in order to find most likely column.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r'c:\Users\rajku\Downloads\archive (1)\Titanic-Dataset.csv')[['Age','Fare', 'Pclass', 'Survived']]
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
x_train , x_test, y_train , y_test = train_test_split(x,y, test_size=.2)
```

```python
knn = KNNImputer(n_neighbors= 5, weights='distance')
x_train_trf = knn.fit_transform(x_train)
x_test_trf = knn.fit_transform(x_test)
```

```python
pd.DataFrame(x_train_trf , columns=x_train.columns)
```

```python
lr = LogisticRegression()
lr.fit(x_train_trf, y_train)
y_pred = lr.predict(x_test_trf)
```

```python
print(accuracy_score(y_test, y_pred))
```
