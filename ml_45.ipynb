```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r'c:\Users\rajku\Downloads\train (1).csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head(15)
```

```python
df.dropna(inplace=True)
x = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
print(df.isnull().sum())
print(df.isnull().sum().sum())
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=21))
```

```python
x['familiy_size'] = x['Parch'] + x['SibSp'] +1
x
```

```python
def funct(num):
    if num ==1:
        return 0
    elif num >1 and num <=4:
        return 1
    else:
        return 2
```

```python
x['family_type'] = x['familiy_size'].apply(funct)
```

```python
x
```

```python
x.drop(columns=['SibSp','Parch','familiy_size'], inplace=True)
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```
