## Regression Metrics

MAE (mean absolute error)
MSE (mean square Error)
RMSE (root mean square error)
R2 Score
adjusted R2 Score

### For calculating MAE)
* Algprithm -> make a Regression line -> Distance btw line and point (Absolute error) ->
### MAE =  |Y1'-Y1| + |Y2' - Y2|...../ n
* We need to minimize the value of Mean absolute error

* MAE comes in terms of Y, makes it easy to understand, 
* Robust to outlier

### MSE = |Y1' - Y1|^2 + |Y2' - Y2|^2..../n
* MSE is differentiable (advt)
* Outlier will be impacted very much (disadv)

### RMSE  = sqroot(MSE)
Used mostly in deeplearning 


### R2 Score
* The R² (R-squared) score, or coefficient of determination, is a statistical metric used in machine learning to evaluate the performance of a regression mode
* R^2 = 1- (SS)res/ (SS)mean. 
1. SSres = square error in regresssion line
2. SSmean = square error in meamline


### Adjusted R2 Score:
#### Radj = 1 - [(1-R2)(n-1) / (n-1-k)]
n = no. of rows
k = independent (1,2,3..)
eg. new columns as temprature, which is irrelevent column, shouldnt affact the score but adj R2 score decrease
