## Gradient Descent
#### Graient means derivative
* Gradient descent is a fundamental mathematical optimization algorithm used to minimize a cost function
* It works by taking iterative steps in the direction of the steepest descent (the negative gradient) to find the optimal model parameters (like weights and biases) that yield the smallest possible error.
* used in Linear reg, Logistic Reg, Tsne (Deep learning backbone)

### 1 Batch Gradient Descent (BGD)
* How it works: Processes the entire training dataset in one single step to calculate the gradient and update the model parameters.
* Pros: Smooth, stable path to the global minimum.Cons: Extremely slow and highly memory-intensive for large datasets.
* Best for: Small, structured datasets where computational power isn't a bottleneck

### 2 Stochastic Gradient Descent (SGD / Stukasting GD)
* How it works: Processes one random training example at a time to update the model parameters.
* Pros: Highly memory-efficient and much faster for massive datasets.
* Cons: The path to the minimum is very noisy and erratic, which means the model may constantly "bounce around" without fully stabilizing.
* Best for: Real-time learning, streaming data, and escaping local minima.

 ### 3 Mini-Batch Gradient DescentHow
 * it works: Divides the dataset into small, manageable subsets (commonly called batches of 32, 64, or 128) and updates the parameters for each subset.
 * Pros: Combines the best of both worlds—stable convergence and computational efficiency (especially optimized for GPUs).
 * Cons: Requires tuning to select the optimal batch size.
 * Best for: The default standard used across almost all deep learning and large-scale model training today.

```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x, y = make_regression(n_features=1, n_samples=4 , noise=80, n_targets= 1, n_informative=1)
```

```python
plt.scatter(x,y)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x, lr.predict(x), color='red')
plt.show()
```

```python
y_pred = ((122.7*x) + 0).reshape(4)
# for slope 122.7 we hav blue line

plt.scatter(x,y)
plt.plot(x, lr.predict(x), color='red')
plt.plot(x,y_pred,color='b')
plt.show()
```

```python
m = 122.7
b = 0
# formula = -1 sum(Yi - m*Xi - b)
loss_slope = -2* np.sum(y - m*x.ravel() - b )
#  ravel() function
loss_slope
```

```python
# lets take learning rate = 0.1
lrr = 0.1

step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((122.7*x) + b).reshape(4)

plt.scatter(x, y, label='Data points')
plt.plot(x, lr.predict(x), color='b', label='Predicted (lr)')
plt.plot(x, y_pred, color='red', label='b = 0')
plt.plot(x, y_pred1, color='green', label='b = 17.9')
plt.legend() #legend is used to for lables showing

plt.plot()

# we can see green is almost matching to the blue(predicted)
```

```python
# Iteration 2
loss_slope = -2* np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lrr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((122.7*x) +b ).reshape(4)
y_pred2
```

```python
plt.scatter(x, y, label='Data points')
plt.plot(x, lr.predict(x), color='b', label='Predicted (lr)')
plt.plot(x, y_pred, color='red', label='b = 0')
plt.plot(x, y_pred1, color='green', label='b = 17.9')
plt.plot(x, y_pred2, color='orange', label='b = 25.09')
plt.legend() 

plt.plot()
```
