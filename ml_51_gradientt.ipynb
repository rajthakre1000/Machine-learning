```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
x,y = make_regression(n_features=1, n_samples=100 , n_informative=1, n_targets=1 , noise=20 )
```

```python
plt.scatter(x,y)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
m = 32.8
```

```python
class GDRegressor :
    def __init__(self, learning_rate, epochs):
        self.m = 32.8
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs
    def fit(self,x,y):
        # calculating b using GD
        history =[]
        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y - self.m*x.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope)
            print(loss_slope, self.b)
            history.append({'loss_slope': loss_slope, 'b_intercept': self.b})
            
        print(self.b)
        
        # 3. Return the collected data
        return history
        print(self.b)
```

```python
gb = GDRegressor(0.001, 50)
```

```python
gb.fit(x,y)
```
